```python
# Bivariate / Multivariate
'''
# Bivariate:-Bivariate analysis studies the relationship between two variables at a time.
Bivariate Analysis – Examples from Tips Dataset
Example 1: Do smokers tip more?

Variables used (2 variables):

smoker (categorical)

tip (numerical)

Type:
Categorical vs Numerical → Bivariate


Multivariate:-
Multivariate Analysis – Examples from Tips Dataset
Example 4: Do smokers tip more AFTER considering total bill?
Variables used (3 variables):
tip → Dependent variable
smoker
total_bill

Type:
Multivariate analysis
Why multivariate?
Smokers may tip more only because they have higher bills
We control for total_bill

Interpretation:
After controlling for bill size, smokers do not tip significantly more

Conclusion (Multivariate):
The apparent difference in tips is mostly due to bill amount, not smoking status
'''
```

```python
# Do  smokers tip more?
tips_more=df.groupby('smoker')['tip'].mean()
print(tips_more)
```

```python
# Does table size affect the bill proportionally?
avg_bill_by_size = df.groupby("size")["total_bill"].mean()
print(avg_bill_by_size)
```

```python
sns.set_style('whitegrid')
# sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time
plt.figure(figsize=(10,6)) #10:6 ratio figure size
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', s=100) # data
plt.title('The relationship between total bill and tip' , fontsize=16, pad=20)
plt.xlabel('Total Bill amount($)',fontsize=12)
plt.ylabel('Tip amount $',fontsize=12)
plt.show()
```

**Bar Graph**

```python
# Bar Graph:- Comparison
# which day of the week generates highest average total bill
plt.figure(figsize=(10,6))
sns.barplot(data=tips, x='day', y='total_bill',palette='viridis',capsize=.1)
plt.title('Average total bill by day of the week',fontsize=16, pad=20)
plt.xlabel('Day of the week',fontsize=12)
plt.ylabel('Average total bill ($)',fontsize=12)
plt.show()
```

# **Line Graph**

```python
# As table size increases ,How does the average tip amount change?
plt.figure(figsize=(10,6))
sns.lineplot(data=tips, x='size', y='tip', hue='time', palette='viridis')
```

```python
# from matplotlib.lines import lineStyles
plt.figure(figsize=(10,6))
sns.lineplot(data=tips, x='size',y='tip',marker='o',color='teal',linewidth=2.5)
plt.title('Tip Trend based on table size',fontsize=15)
plt.xlabel("Number of people at table (size)",fontsize=12)
plt.ylabel("Average Tip ($)",fontsize=12)
# That is adding grid for ability
plt.grid(True,linestyle='--',alpha=0.6)
plt.show()
```

```python
# multivariate line plot
# How
plt.figure(figsize=(12,6))
sns.lineplot(data=tips, x='size',y='tip',hue='time',style='time',markers=True,linewidth=3)
plt.title('Tip Trend:Lunch vs Dinner by Table Size',fontsize=15,fontweight='bold')
plt.xlabel("Number of people at table (size)",fontsize=12)
plt.ylabel("Average Tip ($)",fontsize=12)
plt.legend(title='Shift Time',title_fontsize='13',fontsize='11')
# That is adding grid for ability
plt.grid(True,linestyle=':',alpha=0.7)
plt.show()
```

# **Box Plot**

```python
# how does the distribution of the total bill vary across the different days ?
# Are there more big-spenders on weekends?
plt.figure(figsize=(12,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='day',palette='Set2',linewidth=2)
plt.title('Distribution of total Bills by Days',fontsize=16)
plt.xlabel('Day of the week',fontsize=12)
plt.ylabel('Total Bill ($)',fontsize=12)
plt.legend([],[],frameon=False)
plt.show()
```

# **Histogram and Density **

```python
# Histogram:-count how many
# kde:how lightly
# what is the most common bill amount at this restaurant and how skewed is the spending behaviour?
plt.figure(figsize=(12,6))
sns.histplot(data=tips,x='total_bill',kde=True,bins=30,color='midnightblue',alpha=0.4)
plt.title('Distribution of total_bills (Histogram and KDE)',fontsize=16)
plt.xlabel('Total Bill amount',fontsize=12)
plt.ylabel('Density',fontsize=12)

plt.axvline(tips['total_bill'].mean(),color='red',linestyle='--',label='Mean')
plt.grid()
plt.legend()
plt.show()
```
