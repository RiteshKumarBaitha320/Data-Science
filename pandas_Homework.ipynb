```python
# How does job satisfaction vary based on the number of working hours per week? Hint: JobSat and WorkWeekHrs
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Tech Axis/survey_results_public (1).csv')
df.head()
```

```python
# How does job satisfaction vary based on the number of working hours per week? Hint: JobSat and WorkWeekHrs
df.groupby('JobSat')['WorkWeekHrs'].mean()
```

```python
df.groupby('JobSat')['WorkWeekHrs'].mean().plot(kind='bar')
```

```python
# What are the most common web frameworks used by developers, and are there any preferences for the next year? Hint: WebFrameWorkedWith and WebFrameDesireNextYear
df.groupby('WebFrameWorkedWith')['WebFrameDesireNextYear'].value_counts()
```

```python
# Is there any relationship between compensation and the number of years of coding experience? Hint: ConvertedComp
df.groupby("YearsCode")["ConvertedComp"].mean()
```

```python
# Do developers who work remotely report higher career satisfaction compared to those who don't? Hint: WorkRemote and JobSat
df.groupby('WorkRemote')['JobSat'].value_counts(normalize=True)
```

```python
# What is the average age of developers across different countries, and how does it affect their career satisfaction? Hint: CareerSat
df.groupby('Country')['Age'].mean()


```

```python
print("------------------Age affect their carrier satisfactions------------------")
df.groupby("CareerSat")["Age"].mean()
```

```python
#Education Data Analysis
# Get the EdLevel data of Nepal
# Count and plot the EdLevel data of Nepal
df[df['Country']=='Nepal']['EdLevel'].value_counts()
```

```python
df[df['Country']=='Nepal']['EdLevel'].value_counts().plot.bar()
```

```python
# Perform grouping of countries and find the EdLevel data
df.groupby('Country')['EdLevel'].value_counts()
```

```python
# Compare the education status of developers from India and Nepal.
df_filtered = df[df['Country'].isin(['Nepal', 'India'])]
```

```python
df_filtered.groupby('Country')['EdLevel'].value_counts()
```

```python
# Using agg function, compute the mean, median, max, and min data of age of all countries.
df.groupby('Country')['Age'].agg(['mean', 'median', 'max', 'min'])
```
