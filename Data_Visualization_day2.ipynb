```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5]
temperature=[22,24,24,26,28]
```

```python
import matplotlib.pyplot as plt
plt.plot(days,temperature)
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("Daily Temperature Graph")
plt.show()
```

```python
hour=[1,2,3,4,5,6]
study_time=[2,3,4,7,9,1]
plt.plot(hour,study_time)
plt.xlabel("hour")
plt.ylabel("study_time")
plt.title("study time")
plt.show()
```

```python
# Bar chart
#  stationery shop to identify total sales
items=['Pen','Book','Pencil','Ruler']
sales=[50,305,400,100]
plt.bar(items,sales)
plt.xlabel('items')
plt.ylabel('sales')
plt.title('Stationery Shop')
plt.show()
```

```python
sub=['English','Math','Science']
marks=[90,80,70]
plt.bar(sub,marks)
plt.xlabel('subject')
plt.ylabel('marks')
plt.title('Student Marks')
plt.show()
```

```python
# scatter plot
sub=['English','Math','Science']
marks=[90,80,70]
plt.scatter(sub,marks)
plt.xlabel('subject')
plt.ylabel('marks')
plt.title('Student Marks')
plt.show()
```

```python
# scatter plot
hour=[1,2,3,4,5]
marks=[90,80,70,60,50]
plt.scatter(hour,marks)
plt.xlabel('hour')
plt.ylabel('marks')
plt.title('Student Marks')
plt.show()
```

```python
age=[10,20,30,40,45]
height=[100,120,130,140,150]
plt.scatter(age,height)
plt.xlabel('age')
plt.ylabel('height')
plt.title('Age wise Height Difference')
plt.show()
```

```python
# Multiple line comparison
days=[1,2,3,4,5]
student1=[40,50,60,70,80]
student2=[40,45,55,60,70]
plt.plot(days,student1)
plt.plot(days,student2)
plt.bar(days,student1)
plt.bar(days,student2)
plt.xlabel('days')
plt.ylabel('student')
plt.title('Student report Comparison')
plt.legend(['student1','student2'])
plt.show()
```
